### **FILLING NULL VALUES**

####**FILLING THE NUMERICAL COLUMN WITH THEIR MEAN**

```python
#Filling the numerical column with their mean
df['bp'].fillna(df['bp'].mean(),inplace=True)
df['sg'].fillna(df['sg'].mean(),inplace=True)
df['al'].fillna(df['al'].mean(),inplace=True)
df['su'].fillna(df['su'].mean(),inplace=True)
df['bgr'].fillna(df['bgr'].mean(),inplace=True)
df['bu'].fillna(df['bu'].mean(),inplace=True)
df['sc'].fillna(df['sc'].mean(),inplace=True)
df['sod'].fillna(df['sod'].mean(),inplace=True)
df['pot'].fillna(df['pot'].mean(),inplace=True)
df['hemo'].fillna(df['hemo'].mean(),inplace=True)
df['hemo'].fillna(df['hemo'].mean(),inplace=True)
df['hemo'].fillna(df['hemo'].mean(),inplace=True)
df['pcv'].fillna(df['pcv'].mean(),inplace=True)
df['wc'].fillna(df['wc'].mean(),inplace=True)
df['rc'].fillna(df['rc'].mean(),inplace=True)
```

```python
df = df[df['age'].notna()]
```

```python
df['age'].isnull().sum()
```

```python
df.shape
```

####**FILLING THE CATEGORICAL COLUMN WITH THEIR MOST OCCURRING CLASS**

```python
df = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
df.isnull().sum()
```

```python
categorical_columns=[]
for col in df.columns:
        if df[col].dtype=='object':
            categorical_columns.append(col)
```

```python
for i in categorical_columns:
  print('{} values :{}'.format(i,df[i].unique()))
```

### **REMOVING \t in CATEGORICAL COLUMNS**

```python
df['dm'] = df['dm'].replace(r'\t',' ', regex=True)
df['dm'] = df['dm'].str.strip()
df['cad'] = df['cad'].replace(r'\t',' ', regex=True)
df['cad'] = df['cad'].str.strip()
df['classification'] = df['classification'].replace(r'\t',' ', regex=True)
df['classification'] = df['classification'].str.strip()
```

### **VIEWING THE CATEGORICAL COLUMN VALUES AFTER TRIMMING AND REMOVING '\t'**

```python
for i in categorical_columns:
  print('{} values :{}'.format(i,df[i].unique()))
```
